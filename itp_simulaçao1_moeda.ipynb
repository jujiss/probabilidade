# Experimentos

```python
import numpy as np 

p = 0.4 # probabilidade de cair cara
n = 11  # quantidade de vezes que a moeda foi jogada
c = 0   # caras já obtidas 
r = []  # resultados obtidos, sendo cara=1 e coroa=0

for i in range(0, n): # criação de loop com início em 0 e fim em n
  if (np.random.uniform() < p): # quando o valor obtido for menor que p, cai cara 
      c = c + 1
      r.append(1) # adiciona 1 à r
  else:
      r.append(0) # adiciona 0 à r

print("r= ", r)
print("frequência de caras= ", c/n)
```

# Distribuição e criação de gráfico

```python
# bibliotecas utilizadas

import numpy as np 
import matplotlib.pyplot as plt


# probabilidade de cair cara
p = 0.4   
# máximo de simulações realizadas
n = 1000  
# número de simulações realizadas
sim = []  
# frequência de caras
prob = [] 


# realiza simulações com valores de 10 em 10 de 1 até n
for nsim in range(1,n,10):
# caras já obtidas
  c = 0 
# aumenta a quantidade de simulações realizadas
  for i in range(1,nsim): 
 # quando o valor obtido for menor que p, cai cara
    if(np.random.uniform() < p): 
      c = c + 1
# adiciona razão entre as caras obtidas e o total de simulações realizadas à prob
  prob.append(c/nsim) 
# adiciona o valor de simulações realizadas ao total de simulações 
  sim.append(nsim) 


# tamanho da figura (comprimento, altura)
plt.figure(figsize=(10,7)) 

# especificações da linha da simulação
plt.plot(sim, prob, linestyle='-', color='darkorange', linewidth=3, label= 'simulação') 

# especificações da linha teórica
plt.axhline(y=p, color='black', linestyle='dotted', label='teoria') 

# nome do eixo x
plt.xlabel("Quantidade de experimentos", fontsize=20) 

# nome do eixo y
plt.ylabel("Razão de caras", fontsize=20) 

# limitação do eixo x
plt.xlim([0.0, n]) 

# limitação do eixo y
plt.ylim([0.0, 1.0]) 

# legenda das linhas
plt.legend() 

plt.show(True)
```
