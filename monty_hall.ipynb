# Problema de Monty Hall

### https://pt.wikipedia.org/wiki/Problema_de_Monty_Hall

O problema de Monty Hall foi baseado em um programa de televisão estadunidense, o qual possuía um jogo, descrito abaixo.

Há três portas, sendo que atrás de uma delas está um carro, e atrás das outras duas estão bodes. O jogador escolhe uma das portas, sem saber o que há por trás dela. Posteriormente, o apresentador do programa, que sabe o que há em cada uma das portas, abre uma das não escolhidas, revelando, assim, um bode. Por último, o apresentador questiona o jogador se ele quer permanecer com a sua escolha, ou prefere escolher a outra porta que também não foi aberta ainda. 

O problema está em saber qual é a melhor estratégia: continuar com a porta escolhida inicialmente ou mudar sua escolha. Abaixo, mostro a probabilidade de sucesso no caso em que o jogador escolher mudar a porta.



```python
# numpy: biblioteca para tratar operações matemáticas de maneira vetorial
import numpy as np

# matplotlib: biblioteca para criar simulações visuais de operações matemáticas
from matplotlib import pyplot as plt

# random: biblioteca parar criar sequências pseudo-aleatórias de números
import random 

# seed: dá a entrada para o gerador de números aleatórios
from random import seed 
seed(3)
```

```python
# numeração de cada uma das portas envolvidas no jogo
p = [1,2,3]
# máximo de simulações realizadas
n = 3000
# fração de sucessos (carro atrás da porta)
s = []
# quantidade de jogos realizados
j = []

# sucessos e jogos iniciais;
# contabiliza os sucessos e jogos de toda a simulação
tsucesso = 0
tjogos = 0

# valores são gerados em um intervalo de 1 a n, de dois em dois
for i in np.arange(1,n,2):
# sucessos e derrotas já obtidas
  sucesso = 0
  derrota = 0 
# realiza simulações com valores de 0 a i
  for x in range(0,i):
# coloca o carro em uma porta aleatória
    c = random.choice(p)
# escolhe uma porta aleatória
    pa = random.choice(p)
# o carro não está na porta escolhida;
# o apresentador abre uma das outras portas;
# o jogador escolhe trocar a porta=vitória :)
    if(c != pa):
      sucesso = sucesso + 1
# adiciona razão entre vitórias à s
  s.append(sucesso/i)
# adiciona o valor de jogos realizados ao total de jogos
  j.append(i)


# soma dos sucessos obtidos
  tsucesso += sucesso
# soma dos jogos realizados
  tjogos += x


# divide a quantidade de sucessos obtidos pelos jogos realizados
print("razão de sucessos= ", tsucesso/tjogos)
```

# Gráfico


```python
# tamanho da figura (comprimento, altura)
plt.figure(figsize=(10,8))
# especificações da linha da simulação
plt.plot(j, s, linestyle='-', label='simulação da mudança de porta', color='cadetblue')
# especificações da linha teórica
plt.axhline(y=0.66, linestyle='dotted', color='midnightblue', label='teoria da mudança de porta')
# nome do eixo x
plt.xlabel('jogos', fontsize=20)
# nome do eixo y
plt.ylabel('probabilidade de vitória', fontsize=20)
# legenda das linhas
plt.legend(fontsize=20)

plt.show(True)
```
